```python
# Passo 3: Importar as bases de dados de vendas
tabela_total = pd.DataFrame()
for file in file_list:

  #se houver "Vendas" no nome do arquivo, então
 if "Vendas" in file:
   tabela = pd.read_csv(f"/content/drive/MyDrive/Curso Básico de Python/Vendas/{file}")
   tabela_total = tabela_total.append(tabela)


# Passo 4: Tratar / Compilar as bases de dados (todas as bases em uma só)
display(tabela_total)

```

```python
# Passo 5: Calcular o produto mais vendido (em quantidade)

tabela_produtos = tabela_total.groupby('Produto').sum()
tabela_produtos = tabela_produtos[['Quantidade Vendida']].sort_values(by='Quantidade Vendida', ascending=False) #OBS: Dois [], são em casos que precise de mais de uma lista (porém, nesse caso foi usado, para a formatação ficar mais agradável)
display(tabela_produtos)
```

```python
# Passo 6: Calcular o produto que mais faturou (em faturamento)
tabela_total['Faturamento'] = tabela_total['Quantidade Vendida'] * tabela_total['Preco Unitario']

tabela_faturamento = tabela_total.groupby('Produto').sum()
tabela_faturamento = tabela_faturamento[['Faturamento']].sort_values(by='Faturamento', ascending=False)

display(tabela_faturamento)
```

```python

# Passo 7: Calcular a loja/cidade que mais vendeu (em faturamento) - criar um gráfico/dashboard

tabela_lojas = tabela_total.groupby('Loja').sum()
tabela_lojas = tabela_lojas[['Faturamento']].sort_values(by='Faturamento', ascending=False) #sort_values = ordena valores; ascending(crescente) = False (Do maior para o menor) True (Menor para o maior)
display(tabela_lojas)
```

```python
#dashboard feito a partir da biblioteca plotly.express
grafico = px.bar(tabela_lojas, x=tabela_lojas.index, y='Faturamento')
grafico.show()
```
